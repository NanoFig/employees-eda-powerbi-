## 1.3 Conversiones de tipos

```python
#Convertir Start Date a datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], format='%m/%d/%Y', errors='coerce')
```

```python
#Convertir Senior Management a boolean
df["Senior Management"] = df["Senior Management"].astype(bool)
```

```python
#Conversion a categoria
df["Gender"] = df["Gender"].astype("category")
```

```python
# Convertir Last login time a formato solamente en hora y minutos
# Reemplazar los indicadores de AM/PM en español por los estándar en inglés

# En una primera conversión usamos coerce y perdimos el dato original, por eso volvemos a cargar esa columna del dataset original

df_original = pd.read_csv(path, sep = ";", encoding='latin-1')
df["Last Login Time"] = df_original["Last Login Time"]
df["Last Login Time"] = df["Last Login Time"].astype(str).str.strip()

df["Last Login Time"] = pd.to_datetime(
    df["Last Login Time"],
    format="%I:%M %p",
    errors="coerce"
).dt.strftime("%H:%M")

# Verificamos el resultado
df["Last Login Time"].head()
```

```python
df.info()
df.head()
```

# 2. Limpieza del dataset

## 2.1 Estrategia de imputación en faltantes

**Reglas aplicadas:**
- Para First Name sustituir los valores faltantes por "NN"
- Para Gender sustituir los valores faltantes por "No declarado"
- Para Senior Mangament sustiuir valores faltantes por la moda
- Para Team sustituir valores faltantes por la moda  

```python
# Para First Name sustituir los valores faltantes por "NN"
df['First Name'].fillna('NN', inplace=True)
```

```python
# Para Gender sustituir los valores faltantes por "No declarado"

# Primero, agregar "No declarado" si hace falta
if df["Gender"].isnull().any() and "No declarado" not in df["Gender"].cat.categories:
    df["Gender"] = df["Gender"].cat.add_categories("No declarado")

# Reemplazar NaN por "No declarado"
df["Gender"] = df["Gender"].fillna("No declarado")
```

```python
# Para Senior Mangament sustiuir valores faltantes por la moda
df['Senior Management'].fillna(df['Senior Management'].mode()[0], inplace=True)
print(df['Senior Management'].isnull().sum())
print(df['Senior Management'].head(11))
print(df['Senior Management'].mode()[0])
```

```python
#Para Team sustituir valores faltantes por la moda
df['Team'].fillna(df['Team'].mode()[0], inplace=True)
print(df['Team'].isnull().sum())
print(df['Team'].head(11))
print(df['Team'].mode()[0])
```

## 2.2 Corrección de texto
- Normalizamos caracteres, buscar si hay alguna entrada que no coincida con las entradas estandars y modificarlas, eliminar mayúsculas/ minúsculas inconsistentes.



```python
columnas_texto = ["First Name", "Team"]

# Eliminar espacios al inicio/final
for col in columnas_texto:
    df[col] = df[col].str.strip()

# Capitalización
df["First Name"] = df["First Name"].str.title()
df["Team"] = df["Team"].str.title()
```

```python
# Mirar si hay alguna entrada en Gender que no sea Male, Female o No declarado
print(df['Gender'].unique())
# Buscar si todos los First Name inician con mayúscula, en caso de un False normalizar la primera letra a mayúscula
print(df['First Name'].str.istitle().all())
df['First Name'] = df['First Name'].str.title()
print(df['First Name'].str.istitle().all())
```

```python
df.head()
```

## 2.3 Detección y tratamiento de outliers

### Usamos los siguientes criterios para detectar outliers.
- Si Bonus % es menor o igual a 2 lo incrementamos a 5. Si Bonus % es mayor o igual a 15 lo reducimos a 15.
- Para la columna Start Date, analizar si la entrada fue hace más de 30 años (tomando el año actual 2025) entonces el trabajador tiene los años necesarios para jubilarse.

```python
#Si Bonus % es menor o igual a 2 lo incrementamos a 5. Si Bonus % es mayor o igual a 15 lo reducimos a 15.
df.loc[df['Bonus %'] <= 5, 'Bonus %'] = 5
df.loc[df['Bonus %'] >= 15, 'Bonus %'] = 15
# Ahora buscar si hay elemntos en Bonus menores a 5 o mayores a 15
print(df.loc[(df['Bonus %'] < 5) | (df['Bonus %'] > 15)])
```

```python
# Para la columna Start Date, analizar si la entrada fue hace más de 30 años (tomando el año actual 2025) entonces el trabajador debería estar jubilado
df['Años de servicio'] = 2025 - df['Start Date'].dt.year
df['Revisar caso'] = df['Años de servicio'] >= 30
print(df[['Start Date', 'Años de servicio', 'Revisar caso']])
# Si el trabajador debería estar jubilado pero no lo está, marcar para revisar
```

##  3.2 Frecuencias en variables categóricas

```python
df_cat = df.select_dtypes(include=['object','category','boolean'])

# Quitar Last Login Time
if 'Last Login Time' in df_cat.columns:
    df_cat = df_cat.drop(columns=['Last Login Time'])
print(df_cat.columns)

# Analizar la frecuencia de categorías en variables cualitativas.
for col in df_cat.columns:
    print(f"\nFrecuencia de categorías en {col}:")
    print(df_cat[col].value_counts().head(10))
    print()
```

```python
# Countplots de columnas categoricas
import math

# Excluimos nombre
df_cat2 = [c for c in df_cat if 'name' not in c.lower()]

# Calcular número de filas y columnas según cuántas variables haya
n_cols = 2  # número de gráficos por fila
n_rows = math.ceil(len(df_cat2) / n_cols)


# Crear figura general
plt.figure(figsize=(14, n_rows * 4))


# Graficar cada variable categórica
for i, col in enumerate(df_cat2, 1):
    plt.subplot(n_rows, n_cols, i) # Add subplot for correct plotting
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.title(f'Conteo de {col}')
    plt.xlabel('')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

##### *Countplot es un tipo específico de gráfico de barras para variables categóricas que muestra la frecuencia absoluta de cada categoría.*

##  3.3 Tendencia temporal de ingresos por año

```python
#Cantidad de empleados incorporados por año
df['Año de inicio'] = df['Start Date'].dt.year
empleados_por_año = df['Año de inicio'].value_counts().sort_index()
plt.figure(figsize=(10,3))
sns.lineplot(x=empleados_por_año.index, y=empleados_por_año.values, marker='o')
plt.title('Cantidad de empleados incorporados por año')
```

```python
#Para los dos años con mayor cantidad de incorporaciones, mostrar un lineplot con la tendencia mensual dentro de esos años.
top_2_años = empleados_por_año.nlargest(2).index.tolist()
df_top_años = df[df['Año de inicio'].isin(top_2_años)].copy()
df_top_años['Mes de inicio'] = df_top_años['Start Date'].dt.month
empleados_por_mes = df_top_años.groupby(['Año de inicio', 'Mes de inicio']).size().unstack(fill_value=0)
empleados_por_mes.T.plot(figsize=(10,5), marker='o')
plt.title('Tendencia mensual de incorporaciones en los 2 años con más ingresos')
plt.xlabel('Mes')
plt.ylabel('Cantidad de empleados')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend(title='Año')
plt.show()
```

#### *Lineplot muestra la información como una serie de puntos conectados por segmentos de recta, resulta particularmente útil para mostrar tendencias a lo largo de un eje continuo como el tiempo y facilita la comparación, en este caso entre dos años de interés.*

```python
# Incorporaciones por año por team
incorporaciones_team_año = df.groupby(['Año de inicio', 'Team']).size().unstack(fill_value=0)

teams_para_subplot = incorporaciones_team_año.columns[:9]

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

for i, team in enumerate(teams_para_subplot):
    ax = axes[i]
    ax.bar(incorporaciones_team_año.index,
           incorporaciones_team_año[team],
           color=f'C{i}',
           alpha=0.7)

    ax.set_title(f'Equipo: {team}', fontweight='bold')
    ax.set_xlabel('Año')
    ax.set_ylabel('Incorporaciones')
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, alpha=0.3)

# Ocultar subplots vacíos
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Distribución de Incorporaciones por Equipo (Desglose Individual)',
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```

## 4.2 Cruce entre variables categóricas y numéricas

```python
# Cruce entre variables categóricas y numéricas

# Salario por Team
if 'Team' in df.columns and 'Salary' in df.columns:
    salarioxteam = df.groupby('Team')['Salary'].mean().sort_values(ascending=False)
    print("Promedio de salarios por Team:\n")
    print(salarioxteam)
```

```python
# Salario según género
if 'Gender' in df.columns and 'Salary' in df.columns:
    salarioxgenero = df.groupby('Gender')['Salary'].mean()
    print("Promedio de salarios por género:\n")
    print(salarioxgenero)
```

```python
# Relaciones entre variables categóricas y numéricas con boxplot

# Relación Salary y Senior Management
if 'Senior Management' in df.columns:
    print(df.groupby('Senior Management')['Salary'].mean())
    sns.boxplot(x='Senior Management', y='Salary', data=df)
    plt.title('Salary según Senior Management')
    plt.show()
```

```python
# Relación Salary y Team
if 'Team' in df.columns:
    print(df.groupby('Team')['Salary'].mean().sort_values(ascending=False))
    sns.boxplot(x='Team', y='Salary', data=df)
    plt.title('Salary por Team')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Relación Gender y Salary
if 'Gender' in df.columns:
    print(df.groupby('Gender')['Salary'].mean().sort_values(ascending=False))
    sns.boxplot(x='Gender', y='Salary', data=df)
    plt.title('Salary por Gender')
    plt.xticks(rotation=45)
    plt.show()
```

#### *Los boxplots permiten visualizar la distribución de datos numéricos, mostrando forma, dispersión y valores atípicos (outliers). Permite comparar las distribuciones entre los grupos de relevancia, ver si los datos son siméticos o asimétricos y detectar anomalías. *

## 4.3 Cruce entre variables categóricas

```python
# Cruce entre variables categóricas (tabla de contingencia)

if 'Gender' in df.columns and 'Team' in df.columns:
    Teamxgenero = pd.crosstab(df['Gender'], df['Team'], normalize='index').round(2)
    print("Distribución porcentual de Team por Género:\n")
    print(Teamxgenero)
```

```python
# Gráfico de barras apiladas para mostrar porcentaje de Team por Gender

Teamxgenero.plot(kind='bar', stacked=True, figsize=(8,5), colormap='coolwarm')

plt.title('Distribución porcentual de Team por Género')
plt.ylabel('Proporción')
plt.xlabel('Género')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=0)
plt.show()
```

#### *Gráfico de barras apiladas: util para mostrar cómo se componen los totales de interés según categorías, en este caso para ver cómo se distribuyen los empleados de cada género en los equipos. Al mostrar sólo distribución, es adecuado hacer la lectura en paralelo con el gráfico de correlaciones.*

```python
# Heatmap tabla de contingencia

plt.figure(figsize=(10,4))
sns.heatmap(Teamxgenero, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Distribución porcentual de Team por Género\n')
plt.ylabel('Gender')
plt.xlabel('Team')
plt.show()
```

En principio, los principales problemas detectados fueron los valores NaN presentes en la mayoría de las categorías, así como inconsistencias en los tipos de datos.
